```python
import pandas as pd
```

```python
df1 = pd.read_csv("data/piñonesISO.csv")
df2 = pd.read_csv("data/piñonesASA.csv")
df3 = pd.read_csv("data/piñones15.csv")
df4 = pd.read_csv("data/piñones13.csv")
df5 = pd.read_csv("data/piñones14.csv")
df6 = pd.read_csv("data/piñones16.csv")
df7 = pd.read_csv("data/discos.csv")
df8 = pd.read_csv("data/tensores.csv")
df9 = pd.read_csv("data/ruedas.csv")
df10 = pd.read_csv("data/coronas.csv")
```

```python
urls = [
    'https://tracensl.com/familia/7/PI%C3%91ONES-DE-CADENA-DIN8187-NORMA-ISO',
    'https://tracensl.com/familia/101/PI%C3%91ONES-DE-CADENA-DIN-8188-NORMA-ASA',
    'https://tracensl.com/familia/15/PI%C3%91ONES-DE-CADENA-',
    "https://tracensl.com/familia/13/PI%C3%91ONES-DE-CADENA-",
    "https://tracensl.com/familia/14/PI%C3%91ONES-DE-CADENA-",
    "https://tracensl.com/familia/16/PI%C3%91ONES-",
    "https://tracensl.com/familia/1/DISCOS-DE-CADENA",
    "https://tracensl.com/familia/12/PI%C3%91ONES-TENSORES",
    "https://tracensl.com/familia/10/RUEDAS-DE-FUNDICI%C3%93N",
    "https://tracensl.com/familia/31/CORONAS-PARA-DOS-CADENAS-SIMPLES"
]

# Agregar la columna "URL" a cada DataFrame
df1['URL'] = urls[0]
df2['URL'] = urls[1]
df3['URL'] = urls[2]
df4['URL'] = urls[3]
df5['URL'] = urls[4]
df6['URL'] = urls[5]
df7['URL'] = urls[6]
df8['URL'] = urls[7]
df9['URL'] = urls[8]
df10['URL'] = urls[9]
```

```python
df_final = pd.concat([df1, df2, df3, df4, df5, df6, df7, df8, df9, df10], ignore_index=True)
df_final
```

```python
df_final = df_final.fillna("No disponible")
df_final
```

```python
import pandas as pd
import re
# Definir una función para clasificar si es simple, doble o triple
def clasificar_tipo(Nombre):
    if re.search(r'\bSIMPLE\b', Nombre, re.IGNORECASE):
        return 'SIMPLE'
    elif re.search(r'\bDOBLE\b', Nombre, re.IGNORECASE):
        return 'DOBLE'
    elif re.search(r'\bTRIPLE\b', Nombre, re.IGNORECASE):
        return 'TRIPLE'
    else:
        return 'No especificado'

# Aplicar la función para crear una nueva columna llamada 'Tipo'
df_final['SIMPLE/DOBLE/TRIPLE'] = df_final['Nombre'].apply(clasificar_tipo)

df_final
```

```python
def extraer_info_dientes(nombre):
    match = re.search(r'Z(\d+)', nombre)
    if match:
        return f'Z{match.group(1)}'
    else:
        return 'No especificado'

# Aplicar la función para crear una nueva columna llamada 'InfoDientes'
df_final['Dientes'] = df_final['Nombre'].apply(extraer_info_dientes)
```
